```python
def fn(word):
    if not word:
        return 'Not Accepted'
    if not (word[0].islower() and word[0].isalpha()):
        return 'Not Accepted'
    for c in word[1:]:
        if not (c.islower() and c.isalpha()):
            return 'Not Accepted'
    return 'Accepted'
```

```python
alphabet=set(string.ascii_lowercase)
other_inputs = set(chr(i) for i in range(32, 127)) - alphabet
all_symbols = alphabet | other_inputs

from visual_automata.fa.dfa import VisualDFA

transitions = {}
for state in ['q0', 'q1', 'qDead']:
    transitions[state] = {}
    for char in all_symbols:
        if state == 'q0' and char in alphabet:
            transitions[state][char] = 'q1'
        elif state == 'q1' and char in alphabet:
            transitions[state][char] = 'q1'
        else:
            transitions[state][char] = 'qDead'

dfa = VisualDFA(
    states={'q0', 'q1', 'qDead'},
    input_symbols=all_symbols,
    transitions=transitions,
    initial_state='q0',
    final_states={'q1'},
)
```

```python
dfa.show_diagram()
```
