```python
import pandas as pd
df=pd.read_csv('word_freq.csv')
df.head()
```

```python
sort_df=df.sort_values(by='freq',ascending=False)
sort_df.head()
```

```python
from matplotlib import font_manager as fm, rcParamsDefault
font_path = r"C:\Users\dubey\Downloads\Nirmala.ttf"
prop = fm.FontProperties(fname=font_path)
```

```python
print(sort_df.shape)
print(sort_df.head())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 6))
plt.bar(sort_df["word"][:100].astype(str), sort_df["freq"][:100])  

plt.xticks(
    ticks=range(len(sort_df["word"][:100])), 
    labels=sort_df["word"][:100].astype(str), 
    rotation=90, 
    fontsize=10,
    fontproperties=prop
)

plt.xlabel("Word")
plt.ylabel("Frequency")
plt.yscale("log")
plt.title("Top 100 Word Frequencies")
plt.tight_layout()
plt.show()
```

```python
# remove stop words
threshhold=1e6
sort_df.drop(sort_df[sort_df["freq"] < threshhold].index, inplace=True)
```
